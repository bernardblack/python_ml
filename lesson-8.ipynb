```python
import pandas as pd

col_names = ['Rank', 'Name', 'Platform', 'Year', 'Genre', 'Publisher', 'NA_Sales', 'EU_Sales', 'JP_Sales', 
             'Other_Sales', 'Global_Sales']
```

```python
# Загружаем данные из dataset.csv в DataFrame

df = pd.read_csv('dataset.csv', names=col_names, skiprows=1)
df
```

```python
# Выводим заданное количество идущих подряд строк из любой 
# заданной позиции датафрейма.

def get_part(df, start, count):
    return df.iloc[range(start, start + count)]

get_part(df, 500, 10)    
```

```python
# находим все уникальные жанры 

genres = pd.unique(df['Genre'])
print(genres)
```

```python
# разделяем загруженный датафрейм по жанровым датафреймам

frame = {}
for g in genres:
    frame[g] = df[df['Genre'] == g]

frame['Shooter']
```

```python
# Упорядочиваем строки по возрастанию значений в столбце Year

frame2 = {}
for g in genres:
    frame2[g] = frame[g].sort_values('Year')
frame2['Shooter']
```

```python
# в каждом из полученных жанровых датафреймов производим 
# переиндексацию по возрастанию индекса не меняя порядок строк

frame3 = {}
for g in genres:
    frame3[g] = frame2[g].reset_index()
    
frame3['Shooter']
```

```python
# Сохраняем жанровые датафреймы в csv-файлы

for g in genres:
    frame3[g].to_csv(g+'.csv')
```
