```python
import numpy as np
import scipy.integrate as sci
import matplotlib.pyplot as plt
```

```python
# класс, выполняющий математическое описание криволинейных трапеций

class Trapezoid():
    
    def __init__(self, func, left, right):
        
        self.func = func
        self.left = left
        self.right = right
        
    # площадь криволинейной трапеции    
    def get_area(self):
        
        return sci.quad(self.func, self.left, self.right)[0]
    
    # построение графика функции f(x) в заданном диапазоне
    def plot(self):
        
        plt.figure(figsize=(10, 5))
        
        x = np.linspace(self.left, self.right)
        x2 = np.linspace(self.left - 0.3, self.right + 0.3)
        
        plt.plot(x2, self.func(x2))
        
        plt.fill_between(x,self.func(x),color='yellow', hatch='\\', alpha=0.1)
      
        plt.vlines((self.left, self.right), 0, (self.func(self.left),self.func(self.right)), colors='r')
        
        plt.show()
```

```python
t1 = Trapezoid(lambda x: np.sin(x) + np.cos(2*x), 0, 1.2)
t2 = Trapezoid(lambda x: x*np.tan(x) + 2, 0.2, 0.8)
t3 = Trapezoid(lambda x: x**3 - 10*x**2 + 170, 0, 10)
```

```python
t1.get_area()
```

```python
t2.get_area()
```

```python
t3.get_area()
```

```python
t1.plot()
```
