```python
import csv
import matplotlib.pyplot as plt

x = []
y = []

# Load data from file
with open('dataset.txt') as data:
    rows = csv.reader(data, delimiter = ' ')
    next(rows)
    for row in rows:
        x.append(float(row[0]))
        y.append(float(row[1]))
```

```python
# Plot linear chart

max_y = max(y)
max_x = x[y.index(max_y)]

plt.figure(figsize=(16, 10))
plt.annotate('Maximum', xy=(max_x, max_y), 
             arrowprops=dict(arrowstyle='->'), 
             fontsize = 14, xytext=(max_x + 20, max_y), 
             textcoords='data', horizontalalignment='right', verticalalignment='top')
plt.title('y = f(x)', fontsize=14)
plt.grid(True)
plt.xlabel('x-axis', fontsize = 14)
plt.ylabel('y-axis', fontsize = 14)
plt.plot(x, y)
plt.show()
```

```python
# Step plot

plt.figure(figsize=(16, 10))
plt.step(x, y, where='mid')
plt.title('Step', fontsize=14)
plt.grid(True)
plt.xlabel('x-axis', fontsize = 14)
plt.ylabel('y-axis', fontsize = 14)
plt.show()
```

```python
# Stem plot

plt.figure(figsize=(16, 10))
plt.stem(x, y, use_line_collection=True)
plt.title('Stem', fontsize=14)
plt.grid(True)
plt.xlabel('x-axis', fontsize = 14)
plt.ylabel('y-axis', fontsize = 14)
plt.show()
```

```python
# Bar plot

plt.figure(figsize=(16, 10))
plt.bar(x, y)
plt.title('Bar', fontsize=14)
plt.grid(True)
plt.xlabel('x-axis', fontsize = 14)
plt.ylabel('y-axis', fontsize = 14)
plt.show()
```

```python
# Plot linear charts for y=f(x) and y2=2*f(x)

y2 = [i*2 for i in y]

plt.figure(figsize=(16, 10))
plt.plot(x, y, 'o-', label='original')
plt.plot(x, y2, 'o-.g', label="doubled y")
plt.legend()
plt.grid(True)
plt.title('linear charts for y=f(x) and y2=2*f(x)', fontsize=14)
plt.xlabel('x-axis', fontsize = 14)
plt.ylabel('y-axis', fontsize = 14)
plt.show()
```
