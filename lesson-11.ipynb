```python
import numpy as np
import statistics as st
import scipy.stats as sts
import matplotlib.pyplot as plt
```

```python
df = np.loadtxt('data.txt', dtype=str)
df = np.char.replace(df, ',', '.').astype(np.float64)

print(df)
```

```python
print('mean =', st.mean(df))
print('median =', st.median(df))
print('median_low =', st.median_low(df))
print('median_high =', st.median_high(df))
print('mode =', st.mode(df))
print('pst_dev =', st.pstdev(df))
print('st_dev =', st.stdev(df))
print('pvariance =', st.pvariance(df))
print('variance =', st.variance(df))
print('range =', np.ptp(df))
print('max =', np.amax(df))
print('min =', np.amin(df))
print('cumulative sum =', np.cumsum(df))
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))

ax1.hist(df, bins=30, density=True, label='Гистограмма выборки') 

ax1.legend()
ax1.grid()

df = np.sort(df)

y = np.array([len(df[df <= i]) for i in df]) / len(df)
ax2.plot(df, y, label = 'Функция распределения выборки')

ax2.legend()

plt.show()
```
